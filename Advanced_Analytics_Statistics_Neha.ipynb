```python
import pandas as pd
import numpy as np
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

# Z-test

```python
df = pd.read_csv('passport.csv')
df.head()
```

```python
mu = 30
std = 12.5
alpha = 0.05
```

```python
sample_mean = df['processing_time'].mean()

n = len(df)
```

```python
z_stat = (sample_mean - mu) / (std/np.sqrt(n))
```

```python
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
```

```python
print("Sample Mean:", sample_mean)
print("Z statistics:", z_stat)
print("P value:", p_value)
```

```python
if p_value < alpha:
    print("Null hypothesis is rejected.")
    print("Claim is not supported.")
else:
    print("Failed to accept the null hypothesis.")
    print("Clainm is supported.")
```

# Onesample t_test

```python
df = pd.read_csv('bollywoodmovies.csv')
df.head()
```

```python
mu = 500
alpha = 0.05
```

```python
t_stat, p_value = stats.ttest_1samp(df['production_cost'], mu)

print(t_stat)
print(p_value)
```

```python
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("The belief about the average production cost is not correct.")
else:
    print("Fail to reject the null hypothesis.")
    print("The belief about average production cost is correct.")
```

# Two sample t_test

```python
df = pd.read_csv('healthdrink.csv')
df.head()
```

```python
drink = [5.3, 9.0,5.7, 5.5, 5.4]
no_drink =[8.6,5.8, 10.2, 8.5, 6.8]
```

```python
t_stat, p_value = stats.ttest_ind(drink, no_drink)

print(t_stat, p_value)
```

```python
alpha = 0.05

if p_value <= alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

# Paired t_test

```python
df = pd.read_csv('breakups.csv')
df.head()
```

```python
t_stat, p_value = stats.ttest_rel(df['Before_Breakup'], df['After_Breakup'])

print(t_stat, p_value)
```

```python
alpha = 0.05

if p_value <=alpha:
    print("Reject Null Hypothesis")
    print("Alcohol consumption changed after breakup")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significance change in alcohol consumption")
```

# Anova test

```python
df = pd.read_csv('onestop.csv')
df.head()
```

```python
sales_0 = df['discount_0']
sales_1 = df['discount_10']
sales_2 = df['discount_20']
```

```python
f_stat, p_value = stats.f_oneway(sales_0, sales_1, sales_2)

print(f_stat, p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Discount has a significant impact on average sales quantity.")
else:
    print("Discount does not have a significant impact on average sales quantity.")
```

# Chisquare test

```python
observed = [190, 185, 90, 35]
expected = [175, 200, 100, 25]
```

```python
chi_stat, p_value = stats.chisquare(f_obs = observed, f_exp = expected)

print(chi_stat, p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null hypothesis.")
    print("Captain Cook's belief is not true.")
else:
    print("Fail to reject Null Hypothesis.")
    print("Captain cook's belief is true.")
```
